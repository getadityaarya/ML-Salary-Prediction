```python
import pandas as pd
import numpy as np

df = pd.read_csv("Salary Data.csv")
```

```python
#datatypes
df.info()
```

```python
# Function to validate float values
def validate_float(value):
    return value if isinstance(value, float) else np.nan

# Function to validate text for gender
def validate_gender(value):
    valid_genders = {"male", "female", "other"}
    return value if isinstance(value, str) and value.lower() in valid_genders else "Unknown"

# Function to validate text fields
def validate_text(value):
    return value if isinstance(value, str) else np.nan

# Applying the validation rules
df['Age'] = df['Age'].apply(validate_float)
df['Gender'] = df['Gender'].apply(validate_gender)
df['Education Level'] = df['Education Level'].apply(validate_text)
df['Job Title'] = df['Job Title'].apply(validate_text)
df['Years of Experience'] = df['Years of Experience'].apply(validate_float)
df['Salary'] = df['Salary'].apply(validate_float)
```

```python
df
```

```python
#null values
df.isnull().sum()
```

```python
# Example: Impute missing values 

df['Age'].fillna(df['Age'].mean(), inplace=True) 
df['Gender'].fillna('Unknown', inplace=True) # Filling categorical values
df['Education Level'].fillna(df['Education Level'].mode()[0], inplace=True)
df['Job Title'].fillna('No Title', inplace=True)
df['Years of Experience'].fillna(0, inplace=True)  # Assuming 0 for missing experience
df['Salary'].fillna(df['Salary'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:\n", categorical_cols)

for col in categorical_cols:
    df[col] = label_encoder.fit_transform(df[col])
```
